# Recommendation Algorithm

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split

user_item_matrix = df_ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)

 #Split data into training and test sets
train_data, test_data = train_test_split(df_ratings, test_size=0.2, random_state=42)

# Create user-item interaction matrices for train and test sets
train_user_item_matrix = train_data.pivot(index='userId', columns='movieId', values='rating').fillna(0)


# Fit the KNN model
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(user_item_matrix.values)

# Choose a user (for example, userId = 1) and find the K most similar users
user_index = 1 
distances, indices = knn.kneighbors(user_item_matrix.iloc[user_index].values.reshape(1, -1), n_neighbors=5)
```

```python
#indices of the nearest neighbors
similar_users_indices = indices.flatten()

# Get movies rated by these users but not rated by the target user
user_rated_movies = set(user_item_matrix.columns[user_item_matrix.iloc[user_index] > 0])
neighbor_rated_movies = set()

for i in similar_users_indices:
    neighbor_rated_movies.update(user_item_matrix.columns[user_item_matrix.iloc[i] > 0])

# Recommend movies not seen by the user
recommended_movies = list(neighbor_rated_movies - user_rated_movies)

recommended_movie_titles = df_movies[df_movies['movieId'].isin(recommended_movies)]
print(recommended_movie_titles[['movieId', 'title']])
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error, make_scorer
# Create user-item interaction matrices for train and test sets
train_user_item_matrix = train_data.pivot(index='userId', columns='movieId', values='rating').fillna(0)
test_user_item_matrix = test_data.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Ensure both matrices have the same columns (movies)
train_user_item_matrix, test_user_item_matrix = train_user_item_matrix.align(test_user_item_matrix, join='outer', axis=1, fill_value=0)

# Fit the KNN model
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(train_user_item_matrix.values)

# Function to predict ratings for the test set
def predict_ratings(user_item_matrix, knn_model, test_set):
    predictions = []
    for _, row in test_set.iterrows():
        user_id = row['userId']
        movie_id = row['movieId']
        if user_id in user_item_matrix.index:
            # Find similar users
            distances, indices = knn_model.kneighbors(user_item_matrix.loc[user_id].values.reshape(1, -1), n_neighbors=5)
            # Get ratings of similar users for the movie
            similar_users_ratings = user_item_matrix.iloc[indices.flatten(), :].loc[:, movie_id].values
            # Exclude zero ratings
            similar_users_ratings = similar_users_ratings[similar_users_ratings > 0]
            if len(similar_users_ratings) > 0:
                # Predict rating as the mean of similar users' ratings
                predictions.append(np.mean(similar_users_ratings))
            else:
                predictions.append(0)  # Default rating if no similar users rated the movie
        else:
            predictions.append(0)  # User not in training data
    return predictions

# Call the prediction function with the aligned matrices
predicted_ratings = predict_ratings(train_user_item_matrix, knn, test_data)

# Evaluate the predictions
actual_ratings = test_data['rating'].values
mae = mean_absolute_error(actual_ratings, predicted_ratings)
rmse = np.sqrt(mean_squared_error(actual_ratings, predicted_ratings))

print(f'MAE: {mae:.4f}, RMSE: {rmse:.4f}')
```

Both MAE and RMSE indicate that your model is not very precise, with average errors greater than 1.

# SVD no Surprise

```python
from sklearn.metrics import mean_squared_error

# Step 1: Create the user-item interaction matrix
user_item_matrix = df_ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0).values

# Step 2: Perform SVD
U, sigma, Vt = np.linalg.svd(user_item_matrix, full_matrices=False)

# Sigma comes out as a vector, convert it to a diagonal matrix
sigma = np.diag(sigma)

# Step 3: Reconstruct the matrix using only the top k singular values (rank-k approximation)
k = 50  # Choose the number of features (rank)
U_k = U[:, :k]  # Keep only the first k columns of U
sigma_k = sigma[:k, :k]  # Keep only the top k singular values
Vt_k = Vt[:k, :]  # Keep only the first k rows of Vt

# Approximate the user-item matrix
approx_user_item_matrix = np.dot(np.dot(U_k, sigma_k), Vt_k)

# Step 4: Predict the ratings
predicted_ratings = pd.DataFrame(approx_user_item_matrix, index=df_ratings['userId'].unique(), columns=df_ratings['movieId'].unique())

# Step 5: Evaluate the model (e.g., using RMSE)
def rmse(actual, predicted):
    # Flatten both matrices to compare
    actual_flat = actual.values.flatten()
    predicted_flat = predicted.values.flatten()
    # Calculate RMSE only for non-zero values in the original matrix (observed ratings)
    non_zero_indices = np.nonzero(actual_flat)
    return np.sqrt(mean_squared_error(actual_flat[non_zero_indices], predicted_flat[non_zero_indices]))

# Get actual ratings matrix for evaluation
actual_ratings = df_ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Calculate RMSE
rmse_value = rmse(actual_ratings, predicted_ratings)
print(f"RMSE: {rmse_value}")
```

# Item based Algorithm

```python
# Create item-user interaction matrices (movies as rows, users as columns)
train_item_user_matrix = train_data.pivot(index='movieId', columns='userId', values='rating').fillna(0)
test_item_user_matrix = test_data.pivot(index='movieId', columns='userId', values='rating').fillna(0)

# Align matrices to ensure the same users in both train and test
train_item_user_matrix, test_item_user_matrix = train_item_user_matrix.align(test_item_user_matrix, join='outer', axis=1, fill_value=0)

# Fit the KNN model on items (movies)
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(train_item_user_matrix.values)
```

```python
def predict_ratings_item_based(train_matrix, knn_model, test_data):
    predictions = []
    for _, row in test_data.iterrows():
        user_id, movie_id = row['userId'], row['movieId']
        
        if movie_id in train_matrix.index:
            # Find similar movies
            _, indices = knn_model.kneighbors(train_matrix.loc[movie_id].values.reshape(1, -1), n_neighbors=5)
            
            # Check if the user_id exists in the train_matrix columns
            if user_id in train_matrix.columns:
                # Get ratings from the user for similar movies
                similar_movies_ratings = train_matrix.iloc[indices.flatten(), train_matrix.columns.get_loc(user_id)]
                
                # Predict rating as the mean of similar movie ratings (excluding zeros)
                non_zero_ratings = similar_movies_ratings[similar_movies_ratings > 0]
                predictions.append(non_zero_ratings.mean() if len(non_zero_ratings) > 0 else 0)
            else:
                # If user_id not in train_matrix columns, assign default rating
                predictions.append(0)
        else:
            predictions.append(0)  # Default rating for unknown movies
            
    print(predictions)
    return predictions
```

```python
# Call the prediction function
predicted_ratings = predict_ratings_item_based(train_item_user_matrix, knn, test_data)
print(predicted_ratings)
```

```python
# Evaluate predictions
actual_ratings = test_data['rating']
mae = mean_absolute_error(actual_ratings, predicted_ratings)
rmse = np.sqrt(mean_squared_error(actual_ratings, predicted_ratings))

print(f'MAE: {mae:.4f}, RMSE: {rmse:.4f}')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# Define a pipeline with preprocessing (scaling) and KNN model
knn_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Scaling step (optional based on your data)
    ('knn', NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=5))
])

# Fit the pipeline on the train data
knn_pipeline.fit(train_item_user_matrix.values)

# Prediction function remains the same
def predict_ratings_item_based(train_matrix, knn_pipeline, test_data):
    predictions = []
    for _, row in test_data.iterrows():
        user_id, movie_id = row['userId'], row['movieId']
        
        if movie_id in train_matrix.index:
            # Find similar movies using the pipeline (which includes scaling + KNN)
            _, indices = knn_pipeline.named_steps['knn'].kneighbors(
                train_matrix.loc[movie_id].values.reshape(1, -1), n_neighbors=5
            )
            
            # Check if user_id exists in the train matrix columns
            if user_id in train_matrix.columns:
                similar_movies_ratings = train_matrix.iloc[indices.flatten(), train_matrix.columns.get_loc(user_id)]
                non_zero_ratings = similar_movies_ratings[similar_movies_ratings > 0]
                predictions.append(non_zero_ratings.mean() if len(non_zero_ratings) > 0 else 0)
            else:
                predictions.append(0)  # Default rating for unknown users
        else:
            predictions.append(0)  # Default rating for unknown movies
            
    return predictions

# Call the prediction function with the pipeline
predicted_ratings = predict_ratings_item_based(train_item_user_matrix, knn_pipeline, test_data)
print(predicted_ratings)
```

```python
#Evaluation with pipeline.
def evaluate_pipeline(predictions, actual_ratings):
    mae = mean_absolute_error(actual_ratings, predictions)
    rmse = np.sqrt(mean_squared_error(actual_ratings, predictions))
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    return mae, rmse

# Predict ratings using the pipeline
predicted_ratings = predict_ratings_item_based(train_item_user_matrix,knn_pipeline, test_data)

# Extract actual ratings from the test data
actual_ratings = test_data['rating'].values

# Evaluate the predictions
mae, rmse = evaluate_pipeline(predicted_ratings, actual_ratings)
```
