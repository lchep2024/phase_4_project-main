# Check for missing values.


```python
print('missing values for df_links:')
print(df_links.isnull().sum())

print("\nmissing values for df_movies:")
print(df_movies.isnull().sum())

print("\nmissing values for df_ratings:")
print(df_ratings.isnull().sum())

print("\nmissing values for df_tags:")
print(df_tags.isnull().sum())
```

Tmdbid has 8 missing values,we can either fill or drop the rows depending on the importance of the column.

```python
#drop rows with Nan in tmdbId
df_links=df_links.dropna(subset=['tmdbId'])
df_links.isnull().sum()
```

```python
#convert floats to integers
df_links['tmdbId']=df_links['tmdbId'].astype(int)
df_ratings['rating']=df_ratings['rating'].astype(int)

print("dtypes for df_links:")
print(df_links.dtypes)

print("\ndtypes for df_ratings")
print(df_ratings.dtypes)
```

```python
#Split the genre column and have them as a list.
df_movies['genres']=df_movies['genres'].apply(lambda x: x.split('|'))
df_movies.head()
```
